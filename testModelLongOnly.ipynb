```python
import os
from workWithAudio import *
import pandas as pd


frameLength = 200
modelPath = 'D:\\SD\\yandexML\\pandModel\\models\\model_new_ExtraTreesClassifier_200_9774.sav'
# modelPath = 'D:\\SD\\yandexML\\pandModel\\models\\model_ExtraTreesClassifier_60_9656.sav'
targetsPath = 'D:\\SD\\yandexML\\pandModel\\data\\test'
resp = 'D:\\SD\\yandexML\\pandModel\\data\\testResults3.tsv'
# targetsPath = 'D:\\SD\\yandexML\\pandModel\\data\\extra'
files = os.listdir(targetsPath)
i = 0
# dfRes = pd.DataFrame(columns=['file'] + [fr"spectr_{jj}" for jj in range(128*frameLength)])
dfProt = []
fNames = []
for file in files:
    i+=1
    if(i%100 == 0):
        print(fr"Обработано {i} аудиозаписей")
    
    newAudio = workWithAudio(targetsPath, file)
    [st, fin] = newAudio.getTiming(newAudio.tact)
    croppedAudio = newAudio.tact.T[st: fin]
    l = len(croppedAudio)

    if(l<frameLength):
        res = np.concatenate([croppedAudio, np.zeros((frameLength-l, 128), float)]).reshape(-1)
        # finSpectrList.append(res)
        r = res
    else:
        print("Длинно")
        r = croppedAudio[:frameLength].reshape(-1)
    dfProt.append(r)
    fNames.append(file)
dfRes = pd.DataFrame(dfProt, columns=[fr"spectr_{jj}" for jj in range(128*frameLength)])
dfRes.insert(0,'file', fNames)
```
